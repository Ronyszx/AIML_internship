```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,1,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    return slope * x+ intercept

mymodel = list(map(myfunc, x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()

```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [14.2,16.4,11.9,15.2,18.5,22.1,14.4,25.1,23.4,18.1,22.6,17.2]
y = [215,315,185,332,406,552,412,614,544,421,441,408]

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    return slope * x+ intercept

mymodel = list(map(myfunc, x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```

```python
# Normality is the assumption taht the underlying residuals are normally distributed, or approximately so.
# # Pearson corelation coefficient --- is defined in statics as the measuremnt of strength relationship between two variables and their association with each other

# slope --- Slope of regression line

# intercept -- intercept of the regression line

# rvalue --  The pearsion correleation coeffient. The square of rvalue is equal to the coeffient of determination

# pvalue --- the p-value of hypthosis test whose null hypthosis is that the slope is zero, using Wald Test with y-distribution of the test 
#             statistic. See alternative above for alternative hypthosis
    
# stderr -- Standerd error of the estimated slope under the assumption of residual normally    
```
